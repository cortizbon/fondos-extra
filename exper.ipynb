```python
import pandas as pd
```

```python
dfs = []
for n in range(2015, 2025):
    df = pd.read_excel('fondos.xlsx', sheet_name=f"{n}", skiprows=4)
    dfs.append(df)

```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df['Código de sector'] = df['CÓDIGO ENTIDAD'].str[:2]
df['Código de entidad'] = df['CÓDIGO ENTIDAD'].str[:5].str.split('-').str.join("")
df['Código de unidad'] = df['CÓDIGO ENTIDAD'].str.split('-').str.join("")
```

```python
df.drop(columns=['SECTOR', 'CÓDIGO ENTIDAD'], inplace=True)
```

```python
import json

with open('dic_entities.json', "r") as f:
    dic_entities = json.load(f) 

with open('dic_sec_ents.json', "r") as f:
    dic_sec_ents = json.load(f) 

with open('dic_sector.json', "r") as f:
    dic_sector = json.load(f) 
```

```python
df['Sector'] = df['Código de sector'].astype(int).astype(str).map(dic_sector)
df['Entidad'] = df['Código de entidad'].astype(int).astype(str).map(dic_entities)
```

```python
info = pd.read_excel('info_unidades.xlsx')
```

```python
info = info.ffill()

for col in info.columns:
    try:
        info[col] = info[col].astype(int)
    except:
        print("No se puede convertir")
```

```python
def funcion(row):
    dato = str(row['CÓDIGO DE ENTIDAD'])
    if len(dato) == 3:
        return f'0{dato}'[:2]
    return dato[:2]
```

```python
info['CÓDIGO DE SECTOR'] = info.apply(funcion, axis=1).astype(int)
```

```python
info.columns = info.columns.str.strip()
```

```python
info['CÓDIGO DE SECTOR'] = info['CÓDIGO DE SECTOR'].astype(str)
info['SECTOR'] = info['SECTOR'].str.capitalize()
info['CÓDIGO DE ENTIDAD'] = info['CÓDIGO DE ENTIDAD'].astype(str)
info['ENTIDAD'] = info['ENTIDAD'].str.capitalize()
info['CÓDIGO DE UNIDAD'] = info['CÓDIGO DE UNIDAD'].astype(str)
info['UNIDAD'] = info['UNIDAD'].str.capitalize()

```

```python
def transform_dict(col1, col2):
    dic = dict(info[[col1, col2]]
               .set_index(col1)
               .to_records())
    return dic

dic_sector = transform_dict('CÓDIGO DE SECTOR', 'SECTOR')
dic_entidad = transform_dict('CÓDIGO DE ENTIDAD', 'ENTIDAD')
dic_unidad = transform_dict('CÓDIGO DE UNIDAD', 'UNIDAD')
```

```python
dic_entidad
```

```python
for dic, name in [(dic_sector, 'dic_sector'),
            (dic_unidad, 'dic_unidad'),
            (dic_entidad, 'dic_entidad')]:
    with open(f'{name}.json', 'w') as f:
        json.dump(dic, f)
```

```python
with open('dic_entidad.json', 'r') as f:
    dic_entidad = json.load(f)

with open('dic_sector.json', 'r') as f:
    dic_sector = json.load(f)

with open('dic_unidad.json', 'r') as f:
    dic_unidad = json.load(f)
```

```python
df['Sector'] = df['Código de sector'].astype(int).astype(str).map(dic_sector) 
df['Unidad'] = df['Código de unidad'].astype(int).astype(str).map(dic_unidad) 
df['Entidad'] = df['Código de entidad'].astype(int).astype(str).map(dic_entidad) 
```
